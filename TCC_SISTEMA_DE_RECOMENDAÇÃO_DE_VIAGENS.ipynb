# TCC SISTEMA DE RECOMENDAÇÃO DE VIAGENS

### O que é o Sistema de Recomendação?

Sistemas de Recomendação  tem como objetivo filtrar o conteúdo que é enviado ao usuário, tentar prever a “nota” ou “preferência” que o usuário daria a um conteúdo para então realizar recomendação baseada em suas preferências. Dessa forma, a experiência de consumo dentro da plataforma se torna mais otimizada e interessante.
O sistema deve aprender o padrão de consumo de conteúdo do usuário com experiências passadas e prever o que o usuário gostaria de consumir no futuro.

Um Sistema de Recomendação é uma classe de aprendizado de máquina que oferece sugestões relevantes ao cliente. Antes do sistema de recomendação, a grande tendência para comprar era aceitar sugestões de amigos. Mas agora o Google sabe quais notícias você vai ler, o Youtube sabe que tipo de vídeos você vai assistir com base em seu histórico de pesquisa, histórico de exibição ou histórico de compra.

Um sistema de recomendação  ajuda uma organização a criar clientes fiéis e construir a confiança deles nos produtos e serviços desejados para os quais vieram em seu site. Os sistemas de recomendação de hoje são tão poderosos que também podem lidar com o novo cliente que visitou o site pela primeira vez. Eles recomendam os produtos que estão em alta ou com alta classificação e também podem recomendar os produtos que trazem o máximo de lucro para a empresa.

Um sistema de recomendação de viagens  é um tipo de sistema de recomendação em que temos que recomendar destinos semelhantes ao turista com base em seu interesse.

# 3° Pré-Processamento dos Dados

### Renomeando Colunas

Vamos renomear as colunas de cada arquivo, pois o nome da coluna contém espaço e letras maiúsculas, então faremos as correções para facilitar o uso.

```python
# Rename de Colunas
travel.rename(columns = {'CodDestino':'Cod','Destino':'destino', 'Tipo':'tipodestino', 'Regiao':'regiao'}, inplace=True)
users.rename(columns = {'User-ID':'user_id', 'Destino-Origem':'origem', 'Idade':'idade'}, inplace=True)
ratings.rename(columns = {'User-ID':'user_id', 'CodDestino':'Cod', 'Destino-Nota':'rating'}, inplace=True)
```

```python
travel
```

```python
#Quantidade de Ratings por Usuários
ratings['user_id'].value_counts()
```

```python
# Destinos que tenham mais de 1 avaliação
x = ratings['user_id'].value_counts() > 1
```

```python
x
```

```python
# Quantidade Usuários que fizeram avaliações
# user_ids
y = x[x].index  
print(y.shape)
```

```python
y
```

#### *REGRA de Negócio*

```python
# Trazendo avaliações somente dos usuários q avaliaram mais de 1 destino
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings
```

```python
# Juntando tabelas (Join ou Merge)
rating_with_travel = ratings.merge(travel, on='Cod')
rating_with_travel.head()
```

```python
#Quantidade de rating dos destinos
number_rating = rating_with_travel.groupby('destino')['rating'].count().reset_index()
```

```python
number_rating
```

```python
#Renomeando coluna
number_rating.rename(columns= {'rating':'number_of_ratings'}, inplace=True)
number_rating
```

```python
# Juntando a tabela de destinos com os Ratings com a tabela de quantidade de ratings por destino
final_rating = rating_with_travel.merge(number_rating, on='destino')
final_rating
```

#### *REGRA de Negócio*

```python
# Filtrar somente destinos que tenham pelo menos 1 avaliações
final_rating = final_rating[final_rating['number_of_ratings'] >= 0]
final_rating.shape
```

```python
# Vamos descartar os valores duplicados, porque se o mesmo usuário tiver avaliado o mesmo destino várias vezes, isso pode dar rúim.
final_rating.drop_duplicates(['user_id','destino'], inplace=True)
final_rating.shape
```

```python
final_rating
```


Iremos transpor os **usuários** em **colunas**, ao invés de linhas, pois as avaliações dadas por eles serão as **variáveis** da máquina preditiva.

```python
final_rating.info()
```

```python
# Transposição de linhas(users_id) em colunas
travel_pivot = final_rating.pivot_table(columns='user_id', index='destino', values='rating')
```

```python
travel_pivot
```

```python
travel_pivot.shape
```

```python
travel_pivot.fillna(0, inplace=True)
```

```python
travel_pivot
```

Preparamos nosso conjunto de dados para modelagem. Usaremos o algoritmo de vizinhos mais próximos (nearest neighbors algorithm), que é usado para agrupamento com base na **distância euclidiana**.*




Mas aqui na tabela dinâmica, temos muitos valores zero e no agrupamento, esse poder de computação aumentará para calcular a distância dos valores zero, portanto, converteremos a tabela dinâmica para a matriz esparsa e, em seguida, alimentaremos o modelo.

*Uma matriz densa armazenada em uma matriz NumPy pode ser convertida em uma matriz esparsa usando a representação CSR chamando o csr_matrix () função.

*No mundo da matemática, a distância mais curta entre dois pontos em qualquer dimensão é chamada de distância euclidiana. É a raiz quadrada da soma dos quadrados da diferença entre dois pontos.

```python
from scipy.sparse import csr_matrix
travel_sparse = csr_matrix(travel_pivot)
```

#4° Criação da Máquina Preditiva

```python
from sklearn.neighbors import NearestNeighbors
import pickle 
model = NearestNeighbors(algorithm='brute')
model.fit(travel_sparse)
model2 = model.fit(travel_sparse)
pickle.dump(model2, open('model2.sav', 'wb'))
```

## Novas Predições

```python
#DestinoRJ
distances, suggestions = model.kneighbors(travel_pivot.iloc[0, :].values.reshape(1, -1))
```
